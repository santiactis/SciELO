```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette(sns.color_palette("Blues_r", 10))

df = pd.read_excel('G:\python\gui\country.xls', header=0)
```

```python
df[[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]] = df[[2011, 2012, 2013, 2014, 2015, 
                                                                           2016, 2017, 2018]].fillna(0)
```

```python
df = df.astype({2011:'int32', 2012:'int32', 2013:'int32', 2014:'int32', 2015:'int32', 2016:'int32', 2017:'int32'
           , 2018:'int32'})
```

```python
df_other = df[df['Cited document type'] == 'other']
df_other.drop('Cited document type', axis=1, inplace=True)
df_other = df_other.reset_index(drop=True)

df_book = df[df['Cited document type'] == 'book']
df_book.drop('Cited document type', axis=1, inplace=True)
df_book = df_book.reset_index(drop=True)

df_journal = df[df['Cited document type'] == 'journal']
df_journal.drop('Cited document type', axis=1, inplace=True)
df_journal = df_journal.reset_index(drop=True)

df_proceedings = df[df['Cited document type'] == 'proceedings']
df_proceedings.drop('Cited document type', axis=1, inplace=True)
df_proceedings = df_proceedings.reset_index(drop=True)

df_thesis = df[df['Cited document type'] == 'thesis']
df_thesis.drop('Cited document type', axis=1, inplace=True)
df_thesis = df_thesis.reset_index(drop=True)
```

```python
plt.figure(figsize=(15,8))
f = sns.barplot(x='Affiliation country', y=2011, data=df_other_top_2011)
f.set_ylabel('Cantidad', fontsize=17)
f.set_xlabel('PaÃ­s', fontsize=17)
f.set_title('Top 10 de paises en citas en tipo de publicaciÃ³n "Otros" \n para el aÃ±o 2011', fontsize=20)
f.set_xticklabels(f.get_xticklabels(),rotation=45, fontsize=15)
f.tick_params(axis='y', labelsize=15)

for p in f.patches:
        f.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', fontsize=15)
```
